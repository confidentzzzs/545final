```python
import os
import re

def numerical_sort_key(s):
    """A sorting key function to sort strings with embedded numbers."""
    return [int(text) if text.isdigit() else text.lower() for text in re.split('(\d+)', s)]

def get_nth_image(folder_path, n):
    # List all files in the folder
    files = os.listdir(folder_path)

    # Filter out only image files (assuming .jpg extension)
    images = [file for file in files if file.endswith('.jpg')]

    # Sort the image files using the custom sort function
    images.sort(key=numerical_sort_key)

    # Check if n is within the range of available images
    if 0 <= n < len(images):
        return images[n]
    else:
        return None


# Example usage
# Replace 'your_folder_path' with the path to your folder
# Get the second image (n = 2)
# Replace 'your_folder_path' with the path to your folder
# Get the second image (n = 2)
folder_path = 'drive/MyDrive/image-sets/MSR-VTT-videoTestVideo_video8282-shot1_1'
image_file = get_nth_image(folder_path, 0)
image_path = os.path.join(folder_path,image_file)
print(image_path)
```
